```python
import torch
import matplotlib.pyplot as plt
from generate import generate
```

```python
vae = torch.load('./vae.pt')
CONDITION = 5
device = 'cuda' if torch.cuda.is_available() else 'cpu'

gen_results = []
for _ in range(10):
    gen_results.append(generate(vae, CONDITION, device))
```

```python
_ = plt.figure(figsize=(20, 8))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(gen_results[i])
plt.show()
```
